**MODELS ACCURACY**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Model names and their accuracies
models = ['BERT', 'ROBERTA', 'DISTILBERT', 'ALBERT', 'XLNET', 'TN']
accuracy = [86, 85, 86, 86, 81, 75]

# Set style and color palette
sns.set_style("whitegrid")
colors = sns.color_palette("coolwarm", len(models))

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=models, y=accuracy, palette=colors)

# Add labels and title
plt.xlabel("Model Name", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)
plt.title("Model Accuracy Comparison", fontsize=16, fontweight='bold')

# Show values on bars
for i, acc in enumerate(accuracy):
    plt.text(i, acc + 0.5, f"{acc}%", ha='center', fontsize=12, fontweight='bold')

# Show the plot
plt.show()
```

**OVERALL COMPARISON**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data for Accuracy, LIME Fidelity, and SHAP Fidelity Scores
data_all = {
    "Model": ["BERT", "RoBERTa", "DistilBERT", "ALBERT", "XLNet", "T5"],
    "Accuracy": [86, 85, 86, 85, 81, 75],
    "LIME_Fidelity": [61, 79, 75, 81, 76, 73],
    "SHAP_Fidelity": [79, 77, 74, 80, 76, 90]
}

# Create a DataFrame
df_all = pd.DataFrame(data_all)

# Melt the DataFrame to make it suitable for a grouped bar plot
df_melted = df_all.melt(id_vars=["Model"], value_vars=["Accuracy", "LIME_Fidelity", "SHAP_Fidelity"],
                        var_name="Metric", value_name="Score")

# Set the plot style
sns.set(style="whitegrid")

# Create a single bar plot for Accuracy, LIME Fidelity, and SHAP Fidelity
plt.figure(figsize=(10, 6))
sns.barplot(x="Model", y="Score", hue="Metric", data=df_melted, palette="Set2")

# Title and labels
plt.title("Model Performance and Explainability Metrics", fontsize=16)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Score", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
